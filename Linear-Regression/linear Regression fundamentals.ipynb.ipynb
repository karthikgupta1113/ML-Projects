```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#generating random datasets
np.random.seed(42)
X=np.random.rand(50,1)*100
Y=3.5*X+np.random.rand(50,1)*20
```

```python
#creating and training linear regression model
model=LinearRegression()#--creates linear regression object
model.fit(X,Y)# based on x ,y values finds the best fit line
```

```python
#predicting Y values
Y_pred=model.predict(X)
```

```python
#visualizing the regression line
plt.figure(figsize=(8,5))
plt.scatter(X,Y,color='blue',label='Data Points')
plt.plot(X,Y_pred,color='red',linewidth=2,label='Regression line')
plt.title('linear regression on random dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
print("Slope (Coefficient):",model.coef_[0][0])
print("Intercept:",model.intercept_[0])
```
